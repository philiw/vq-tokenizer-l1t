```python
import numpy as np
import awkward as ak
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq


```

```python
parquet_file = "/eos/project/f/foundational-model-dataset/samples/production_final/VBFHbb/VBFHbb-NEVENT10000-RS35000001.parquet"
```

```python
table = pq.read_table(parquet_file)
```

```python
# show column names
print("Column names in the Parquet file:")
print(table.column_names)
```

```python
array = ak.from_arrow(table['FullReco_JetPuppiAK4_Eta'])
array
```

```python
# 2nd jet of event 5
array[5, 1]
```

```python
# number of jets in the array
ak.num(array)
```

```python
# print entry lenghts of all columns
for column_name in table.column_names:
    column = table.column(column_name)
    ak_array = ak.from_arrow(column)
    try:
        print(f"Length of column '{column_name}': {ak.num(ak_array,axis=1)}")
    except:
        continue
```

```python
# print datatype of each column
for column_name in table.column_names:
    column = table.column(column_name)
    ak_array = ak.from_arrow(column)
    print(f"Datatype of column '{column_name}': {ak.type(ak_array)}")
```
